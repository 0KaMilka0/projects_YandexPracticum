```python
df = df.dropna(subset=['name']).reset_index(drop=True)
df.name.isna().sum()
```

Как ранее выяснили, по столбцам **critic_score, user_score, rating** достаточно большой процент пропущенных значений. Вполне вероятно, что в большинстве строк пропущены значения всех рассматриваемых столбцов. Проверим, где по одним и тем же играм отсутствуют оба типа отзывов и возрастной рейтинг:

```python
df[(df.critic_score.isna() == True) & (df.user_score.isna() == True) & (df.rating.isna() == True)]
```

Итого 6665 строк, где не заполнены все 3 параметра, при этом общее количество пропусков в столбце **rating** - 6764. Выведем на экран срез, где не заполнены только значения в столбцах **critic_score** и **user_score**:

```python
df[(df.critic_score.isna() == True) & (df.user_score.isna() == True) & (df.rating.isna() == False)]
```

Итого 1338 строк, где пропущены и отзывы пользователей и отзывы критиков. При этом общее количество пропущенных значений в рассматриваемых столбцах 9125 и 8578 соответственно. По итогам проведенного анализа уникальных значений среза с пропусками в данных столбцах нельзя прийти к выводу о наличии каких-либо явных закономерностей пропусков (по определенным платформам/в определенные годы/по определенным жанрам). С учетом изложенного, пропуски могли возникнуть при объединении нескольких датасетов, возможно, в части датасетов такие столбцы отсутствовали.<br>
Что касается заполнения выявленных пропусков, то полагаю, что в столбцах user_score и critic_score самостоятельно их заполнять некорректно. Конечно, по строкам, где есть хотя бы один рейтинг, можно было бы предложить заполнить пропуск на основе имеющегося рейтинга, однако, во-первых, процент их корреляции недостаточно однозначен для такого заполнения (58%), во-вторых, по диаграмме ниже можно увидеть достаточно случаев, когда рейтинги критиков и пользователей не совпадают. Поэтому в случае искусственной замены есть высокий риск исказить данные и прийти к ошибочным выводам. 

```python
# расчет корреляции между отзывами пользователей и отзывами критиков
round(df['user_score'].corr(df['critic_score']), 2)
```

```python
#диаграмма рассеяния, отражающая корреляцию исследуемых выше параметров:
df.plot(kind='scatter',
        x='user_score',
        y='critic_score',
        alpha=0.5,
        title='Корреляция между отзывами пользователей и отзывами критиков',
        figsize=(8,6),
        grid=True);
```

Касательно заполнения пропусков по столбцу **rating**: как следует из документации, в имеющихся данных есть только рейтинги, присвоенные ESRB - негосударственная организация, определяющая возрастные рейтинги для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде. В Европе и Японии свои рейтинговые системы и организации - PEGI и CERO. Возможно, игры, по которым отсутствует рейтинг ESRB, были выпущены в Европе или Японии для внутреннего рынка, или им был не присвоен (либо пока не присвоен) рейтинг ЕSRB.<br>
Поскольку из имеющихся данных мы не можем точно определить возрастной рейтинг игры, заполнять пропуски в рассматриваемом столбце полагаю не совсем корректно, тем более, что недостаток данных в большинстве случае приведет к меньшему искажению, чем их некорректное заполнение. При этом с учетом дальнейших вопросов исследования полагаю правильным заменить пропуски на "Нет данных" - замену произведем в следующем пункте вместе с корректировкой обозначений.

Посмотрим на пропуски в столбце **year_of_release**:

```python
df[df.year_of_release.isna() == True]
```

В столбце year_of_release 269 пропусков, при этом по итогам анализа уникальных значений также нельзя сказать, что год пропущен именно по какому-либо одному параметру. <br>
Что касается заполнения пропусков в указанном столбце, то несмотря на то, что в некоторых названиях игр есть год, это не является гарантией, что год, указанный в названии, равен году релиза. Заполнение с учетом идентичного названия игр тоже не считаю корректным, учитывая наличие перевыпусков игр, их выпуска на разных платформах в разные годы. С учетом этого, а также малого количества пропусков в столбце, полагаю возможным их не заменять.

#### Влияние на продажи отзывов пользователей и критиков. Построение диаграммы рассеяния и подсчет корреляции между отзывами и продажами

Для начала согласно заданию проанализируем данный вопрос в рамках одной платформы. Полагаю возможным выбрать самую популярную платформу PS4. Посчитаем корреляцию и построим диаграмму рассеяния по интересующим нас параметрам:

```python
platforms.query('platform == "PS4"')[['total_sales', 'critic_score', 'user_score']].corr()
```

Не совсем большой, но достаточно ощутимый уровень положительной корреляции между объемом продаж и отзывами критиков. Немного неожиданной стала почти нулевая (и отрицательная) корреляция между объемом продаж и отзывами пользователей. Построим диаграммы рассеяния:

```python
fig, ax = plt.subplots(1, 2, figsize=(14,6))
sns.scatterplot(x='user_score', y='total_sales', data=platforms.query('platform == "PS4"'), ax=ax[0])
sns.scatterplot(x='critic_score', y='total_sales', data=platforms.query('platform == "PS4"'), ax=ax[1])
fig.suptitle('Корреляция продаж и отзывов по платформе PS4')
ax[0].set(xlabel='Отзывы пользователей')
ax[1].set(xlabel='Отзывы критиков')
ax[0].set(ylabel='Объём продаж')
ax[1].set(ylabel='Объём продаж')
sns.set_style('darkgrid')
plt.show()
```

Действительно, если по диаграмме корреляции с отзывами критиков еще заметен тренд вправо-вверх, то по диаграмме корреляции с отзывами пользователей такой тренд отсутствует, можно заметить только совсем слабый тренд отрицательной корреляции (влево-вверх). Таким образом, при выборе игр для продажи не следует ориентироваться на отзывы пользователей, по крайней мере в отношении PS4. **Посмотрим далее зависимости по другим платформам.**

Создадим функцию для построения матрицы корреляции и диаграмм рассеяния, затем применим её поочередно к оставшимся платформам: 

```python
def correlation(name):
    display(platforms[platforms.platform == name][['total_sales', 'critic_score', 'user_score']].corr())
    fig, ax = plt.subplots(1, 2, figsize=(14,6))
    sns.scatterplot(x='user_score', y='total_sales', data=platforms[platforms.platform == name], ax=ax[0])
    sns.scatterplot(x='critic_score', y='total_sales', data=platforms[platforms.platform == name], ax=ax[1])
    fig.suptitle(f'Корреляция продаж и отзывов по платформе {name}')
    ax[0].set(xlabel='Отзывы пользователей')
    ax[1].set(xlabel='Отзывы критиков')
    ax[0].set(ylabel='Объём продаж')
    ax[1].set(ylabel='Объём продаж')
    sns.set_style('darkgrid')
    plt.show()
```

```python
correlation("XOne")
```

```python
correlation("3DS")
```

```python
correlation("PC")
```

```python
correlation("WiiU")
```

#### Анализ общего распределения игр по жанрам: что можно сказать о самых прибыльных жанрах? выделяются ли жанры с высокими и низкими продажами?

Построим столбчатую диаграмму по распределению жанров в рамках исследуемого датасета platforms:

```python
platforms\
.pivot_table(index='genre', aggfunc={'name':'count'})\
.reset_index()\
.sort_values(by='name', ascending=False)\
.plot(kind='bar',
    x='genre',
    y='name',
    title='Распределение игр по жанрам за 2014-2016 гг.',
    rot=50,
    grid=True,
    legend=True,
    figsize=(12,6)
)
plt.xlabel('Наименование жанра')
plt.ylabel('Количество игр, шт.')
plt.show()
```

Итак, абсолютное большинство игр выпускается в жанре action, их почти в 3 раза больше, чем ролевых игр, следующих за ними по популярности. Также достаточно много игр выпускается в жанрах Шутер, Спорт, Приключения, Разное. Интересно, что игры-головоломки занимают последнее место, а находящиеся на предпоследнем месте игры-платформеры выпускаются в 2 раза больше головоломок, и почти в том же количестве, что стратегии и симуляции. <br>
Между тем, объем выпуска игр по жанрам хоть и обусловлен спросом, но все же более результативным для понимания нужд потребителей будет **анализ продаж игр по жанрам:**

```python
platforms\
.pivot_table(index='genre', aggfunc={'total_sales':'sum'})\
.reset_index()\
.sort_values(by='total_sales', ascending=False)\
.plot(kind='bar',
    x='genre',
    y='total_sales',
    title='Распределение продаж по жанрам за 2014-2016 гг.',
    rot=50,
    grid=True,
    legend=True,
    figsize=(12,6)
)
plt.xlabel('Наименование жанра')
plt.ylabel('Объём продаж игр, млн.')
plt.show()
```

Закономерно, что в целом распределение почти аналогично ранее проанализированному распределению выпущенных игр: очень высокие продажи по экшен-играм и шутерам. Достаточно высокий объем продаж по РПГ и спортивным играм. Далее продажи уменьшаются почти в 3 раза.<br>
Также в отличие от общего распределения игр по жанрам, в анализе продаж гонки, файтинги и платформеры популярнее, чем симуляции, приключения и стратегии. Головоломки снова на последнем месте.<br>
Таким образом, при формировании продуктового пула необходимо обязательно учитывать в том числе жанры игр.

### Формирование портрета пользователя каждого региона

#### Самые популярные платформы (топ-5) для пользователя каждого региона (NA, EU, JP)

Для составления портрета пользователя обратимся к актуальному датафрейму df_actual, включающего в себя игры с 2014 по 2016 год. <br>
Первым делом составим функцию, позволяющую вывести столбчатые диаграммы и сводные таблицы по интересующим нас вопросам. 

```python
def countries(criterion, x_title, title):
    regions = ['na_sales', 'eu_sales', 'jp_sales']
    fig, ax = plt.subplots(1, len(regions), figsize=(20,8))
    
    for ind, region in enumerate(regions):
        platforms = (
            df_actual
            .pivot_table(index=criterion, aggfunc={region:'sum'})
            .reset_index()
            .sort_values(by=region, ascending=False)
            .head(5)
        )
        display(platforms)
        
        sns.barplot(
            x=criterion,
            y=region,
            data=platforms,
            ax=ax[ind]
        ).set(
            title=region,
            xlabel=x_title,
            ylabel='Объём продаж'
        )
        ax[ind].tick_params(labelrotation=45)
    fig.suptitle(title)
    sns.set_style('darkgrid')
    plt.show()
```

```python
countries('platform', 'Платформы', 'Объем продаж игр по платформам в регионах')
```

Из анализа выведенных графиков следует, что самой популярной платформой в Европе и Северной Америке является PS4, при этом в Европе объем продаж больше на 30 млн. ## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Cформулируем нулевую и альтернативную гипотезы:<br>
**Нулевая гипотеза**: Средние пользовательские рейтинги платформы Xbox One равны средним пользовательским рейтингам платформы PC.<br>
**Альтернативная гипотеза**: Средние пользовательские рейтинги платформ Xbox One и PC различаются.

Для удобства создадим отдельные датафреймы по каждой платформе и отсечем пустые значения:

```python
XOne = df_actual[(df_actual.platform == 'XOne') & (df_actual.user_score.isna() == False)]
PC = df_actual[(df_actual.platform == 'PC') & (df_actual.user_score.isna() == False)]
print(len(XOne), len(PC))
```

Напишем функцию для вывода статистических характеристик выборок:

```python
def print_statistics(column):
    print(f'Среднее значение равно: {column.mean():.2f}\n\
Медиана равна: {column.median():.2f}\n\
Дисперсия равна: {np.var(column, ddof=1):.2f}\n\
Стандартное отклонение равно: {np.std(column, ddof=1):.2f}')
```

```python
print_statistics(XOne.user_score)
```

```python
print_statistics(PC.user_score)
```

Построим гистограммы для визуального анализа распределения пользовательских рейтингов по выборкам:

```python
ax = XOne.plot(kind='hist', 
              y='user_score', 
              bins=15,
              figsize=(12,8), 
              title='Распределение рейтинга пользователей по платформам XOne и PC', 
              label='XOne');
PC.plot(kind='hist', 
              y='user_score', 
              bins=15,
              figsize=(12,8), 
              label='PC',
              alpha= 0.5,
              grid=True,
              ax=ax)
plt.xlabel('Рейтинг пользователей')
plt.ylabel('Количество игр, шт.')
plt.show()
```

Итак, выборки примерно равные по размеру, со схожим распределением, смещенным влево. По имеющимся данным нет оснований полагать, что разные дисперсии, ввиду чего при расчете статистической разности и p-value значение параметра equal_var оставим по умолчанию:

```python
results = st.ttest_ind(XOne.user_score, PC.user_score)
results.pvalue
```

Поскольку при увеличении объема выборки случайные отклонения компенсируют друг друга, и получить значимое различие в выборках при их однородности менее вероятно, а размер выборок у нас не очень большой и составляет 165 и 122, представляется целесообразным установить уровень статистической значимости в 5%:

```python
alpha = .05
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Таким образом, нам не удалось отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов по платформам Xbox One и PC. Полагаю, что нулевая гипотеза действительно с большей степенью вероятности верна, поскольку ранее мы выяснили, что по данным платформам в большинстве своем выходят одни и те же игры. Логично, если в большинстве случаев пользовательские рейтинги по ним совпадают. 

### Средние пользовательские рейтинги жанров Action и Sports разные

Cформулируем нулевую и альтернативную гипотезы:<br>
**Нулевая гипотеза**: Средние пользовательские рейтинги жанра Action равны средним пользовательским рейтингам жанра Sports.<br>
**Альтернативная гипотеза**: Средние пользовательские рейтинги жанров Action и Sports различаются.

Снова создадим отдельные датафреймы по исследуемым жанрам и посмотрим на размер выборок:

```python
action = df_actual[(df_actual.genre == 'Action') & (df_actual.user_score.isna() == False)]
sports = df_actual[(df_actual.genre == 'Sports') & (df_actual.user_score.isna() == False)]
print(len(action), len(sports))
```

Выведем статистические характеристики:

```python
print_statistics(action.user_score)
```

```python
print_statistics(sports.user_score)
```

Построим гистограммы для анализа распределения:

```python
ax = action.plot(kind='hist', 
              y='user_score', 
              bins=15,
              figsize=(12,8), 
              title='Распределение рейтинга пользователей по жанрам Action и Sports', 
              label='Action');
sports.plot(kind='hist', 
              y='user_score', 
              bins=15,
              figsize=(12,8), 
              label='Sports',
              alpha= 0.5,
              grid=True,
              ax=ax)
plt.xlabel('Рейтинг пользователей')
plt.ylabel('Количество игр, шт.')
plt.show()
```

Помимо явно разного размера выборок, их распределение также различно: по рейтингам игр в жанре Спорт оно более растянуто, что свидетельствует о большем стандартном отклонении и дисперсии. При этом отметим, что обе гистограммы имеют скошенность влево.<br>
С учетом изложенного, при расчете p-value параметр equal_var необходимо указать как False:

```python
results = st.ttest_ind(action.user_score, sports.user_score, equal_var=False)
results.pvalue
```

Ввиду не такого большого размера выборок представляется целесообразным установить уровень статистической значимости в 5%:

```python
alpha = .05
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Таким образом, альтернативная гипотеза о различии средних пользовательских рейтингов жанров Action и Sports подтвердилась. Полагаю, это обоснованно, учитывая вышеприведенное распределение, а также очень высокую популярность жанра Action по сравнению с жанром Sports.
